# **1. Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# **2. Importing Dataset**

```python
grading_data = pd.read_csv("Lesson_4_student_grades_dataset.csv")
```

```python
attendance_data = grading_data["Attendance_Percentage"]
grade_data = grading_data["Grades"]
```

# **3. Visualizing Data**

```python
plt.scatter(attendance_data, grade_data)
plt.xlabel('Independent Variable (attendance)')
plt.ylabel('Dependent Variable (grade)')
plt.title('Scatter Plot of Data')
plt.show()
```

# **5. Preparing Data for Modeling:**

```python
x= grading_data[["Attendance_Percentage"]]
y= grading_data[["Grades"]]
```

# **6. Fitting The Linear Regression Model:**

```python
model = LinearRegression()
model.fit(x,y)
```

# **7. Getting Model Coefficients:**

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope: ", slope)
print("Intercept: ", intercept)
```

# **8. Visualizing the Linear Regression Line:**

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), color='red')
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.title("Linear Regression")
plt.show()
```
