# **1. Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **2. Loading Dataset**

```python
titanic_data = pd.read_csv("titanic_data.csv")
```

```python
titanic_data.head()
```

# **3. Data Preparation**

```python
titanic_frame_without_survived = pd.DataFrame(titanic_data, columns = ["PassengerId", "Pclass","Sex","Age","SibSp", "Parch", "Cabin", "Embarked"])
titanic_frame_with_survived = pd.DataFrame(titanic_data, columns = ["Survived","PassengerId", "Pclass","Sex","Age","SibSp", "Parch", "Cabin", "Embarked"])
survived_data = titanic_data["Survived"]

survived_count = titanic_frame_with_survived.groupby("Survived").count()
survived_survived = titanic_frame_with_survived.groupby("Survived")["Survived"].count()

titanic_report = titanic_frame_with_survived.describe()
print(titanic_report)
```

# **4. Correlation Analysis**

```python
survived_count.plot(kind = 'bar', figsize=(12,6))
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survivability Report')
plt.show()

plt.subplot(1,2,1)
survived_survived.plot(kind = 'bar', figsize=(8,4))
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survivability Report')
plt.show()
```
