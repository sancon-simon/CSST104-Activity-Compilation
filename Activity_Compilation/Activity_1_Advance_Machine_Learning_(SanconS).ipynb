# **Importing libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sne
```

# **Loading Data**

```python
hardware_data = pd.read_csv("hardwareStore.csv")
```

# **Initial Data Analysis**

```python
#basic statt with .describe


hardware_stat = hardware_data.describe()
product_categories = hardware_data.groupby('CATEGORY_NAME')['QUANTITY'].sum()
product_categories_count = hardware_data.groupby('CATEGORY_NAME')['STANDARD_COST'].count()
highest_product_count = product_categories_count.idxmax()
highest_categories_count = product_categories_count.idxmax()

print(hardware_stat)
print(product_categories)
print(highest_product_count)
```

# **Category Analysis**

```python
product_categories.plot(kind = 'bar', figsize=(12,6))
plt.xlabel("Category_Name")
plt.ylabel("Quantity")
plt.title("Price_per_Category")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# **COST AND PRICE ANALYSIS**

```python
product_categories_mean = hardware_data.groupby('CATEGORY_NAME')['LIST_PRICE'].mean()
highest_category_price = product_categories_mean.idxmax()

print(product_categories)

product_categories_mean.plot(kind = 'bar', figsize=(12,6))
plt.xlabel("Category_Name")
plt.ylabel("List_Price")
plt.title("Average_Price_per_Category")
plt.xticks(rotation = 360)
plt.tight_layout()
plt.show()
```

# **Location_Analysis**

```python
product_per_city_quantity = hardware_data.groupby('CITY')['QUANTITY'].sum()
city_quant_max = product_per_city_quantity.idxmax()
print(product_per_city_quantity)
print(city_quant_max)

product_per_city_quantity.plot(kind = 'bar', figsize=(12,6))
plt.xlabel("City")
plt.ylabel("Quantity")
plt.title("Product_Count_per_Country")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
